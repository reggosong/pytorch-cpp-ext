```python
import torch

def smooth_relu_ref(x, alpha=1.0):
    zero = torch.zeros_like(x)
    mid = x * x / (2.0*alpha)
    return torch.where(x<0, zero, torch.where(x>alpha, x, mid))
```

```python
x = torch.tensor([-1., 0.2, 2.0], requires_grad=True)
y = smooth_relu_ref(x, 1.0).sum()
y.backward()
print(x.grad)
```
